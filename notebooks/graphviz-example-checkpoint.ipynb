```python
from graphviz import Graph
import pandas as pd
from pathlib import Path 

exhibitions = Path('./moma/exhibitions/MoMAExhibitions1929to1989.csv')
exhibition_data = pd.read_csv(exhibitions)
```

```python
USA15 = exhibition_data.loc[exhibition_data['ExhibitionTitle'] == '15 Americans']
usa_artists = USA15.loc[exhibition_data['ExhibitionRole'] == 'Artist']
usa_curators = USA15.loc[exhibition_data['ExhibitionRole'] == 'Curator']

artist_names = usa_artists['DisplayName'].tolist()
curator_name = usa_curators['DisplayName'].tolist()
```

```python
artist_names
```

```python
# https://graphviz.readthedocs.io/en/stable/manual.html#jupyter-notebooks

d = Graph("15 Americans", filename='15-americans.gv', engine='neato')

d.node('E', '15 Americans')

for a in artist_names:
    d.node(a)
    d.edge('E', a)
d
```
