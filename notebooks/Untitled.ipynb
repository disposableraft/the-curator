## Louise Bourgeois

Let's see a number of artists who are closest to Louise Bourgeois.

```python
indexes, metrics = model.closest(model['louise_bourgeois'], 5)
model.generate_response(indexes, metrics).tolist()
```

```python
indexes, metrics = model.similar('vincent_van_gogh')
model.generate_response(indexes, metrics).tolist()
```

```python
indexes, metrics = model.similar('art_&_language')
model.generate_response(indexes, metrics).tolist()
```

```python
model.distance('john_cage', 'marcel_duchamp')
```

```python
i, m = model.analogy(pos=['john_cage', 'lawrence_weiner'], neg=['art_&_language'])
model.generate_response(i, m).tolist()
```
